```python
import numpy as np
import matplotlib.pyplot as plt

# Define the start date and number of days for the time series
start_date = np.datetime64('2022-01-01')
num_days = 30

# Create an array of dates with daily frequency
dates = start_date + np.arange(num_days)

# Generate sample data: for example, random values between 0 and 100
values = np.random.rand(num_days) * 100

# Function to compute the moving average using a convolution
def moving_average(data, window_size):
    return np.convolve(data, np.ones(window_size)/window_size, mode='valid')

# Compute a 7-day moving average
ma7 = moving_average(values, 7)

# Plotting the time series and the moving average
plt.figure(figsize=(12, 6))
plt.plot(dates, values, label='Daily Values', marker='o')
# Note: The moving average array is shorter by window_size-1 elements.
plt.plot(dates[6:], ma7, label='7-Day Moving Average', color='red', marker='x')
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time Series Data with 7-Day Moving Average")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Define the start date and number of days for the time series
start_date = np.datetime64('2022-01-01')
num_days = 30

# Create an array of dates with daily frequency
dates = start_date + np.arange(num_days)

# Generate sample data: random values between 0 and 100
values = np.random.rand(num_days) * 100

# Function to compute the moving average using a convolution operation
def moving_average(data, window_size):
    return np.convolve(data, np.ones(window_size) / window_size, mode='valid')

# Compute a 7-day moving average
ma7 = moving_average(values, 7)

# Print the time series data
print("Time Series Data:")
for date, value in zip(dates, values):
    print(f"{date}: {value:.2f}")

# Print the 7-day moving average
print("\n7-Day Moving Average:")
# The moving average array is shorter by window_size - 1 elements
for date, ma_value in zip(dates[6:], ma7):
    print(f"{date}: {ma_value:.2f}")
```
